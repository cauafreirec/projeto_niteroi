```python
# Importação de Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dados de Formulários Preenchidos (Exemplo)
data = {
    'Nome': ['João da Silva', 'Maria Oliveira', 'Carlos Sousa', 'Ana Paula', 'Fernando Rocha'],
    'Idade': [22, 19, 23, 21, 20],
    'Bairro': ['São Domingos', 'São Domingos', 'São Domingos', 'São Domingos', 'São Domingos'],
    'Relação com a UFF': ['Estudante', 'Estudante', 'Funcionário', 'Estudante', 'Estudante'],
    'Impacto da Falta de Água': ['Muito', 'Moderadamente', 'Pouco', 'Muito', 'Moderadamente'],
    'Principal Dificuldade': ['Higiene pessoal', 'Alimentação', 'Estudos/Trabalho', 'Higiene pessoal', 'Alimentação'],
    'Utilidade das Soluções': ['Muito úteis', 'Úteis', 'Pouco úteis', 'Muito úteis', 'Úteis'],
    'Melhora da Situação': ['Sim', 'Em parte', 'Não', 'Sim', 'Em parte'],
    'Comentários': [
        'A distribuição de água poderia ser feita mais rapidamente para evitar longas filas.',
        'A distribuição de alimentos ajudou muito.',
        'A solução de água foi lenta.',
        'Agradeço pela iniciativa, foi fundamental para nos ajudar durante essa crise.',
        'A distribuição poderia ser mais organizada.'
    ]
}

# Criação do DataFrame
df = pd.DataFrame(data)

# Exibir os dados
print(df)

# Configuração geral de estilo
sns.set(style="whitegrid")

# Função para adicionar rótulos nas barras
def add_value_labels(ax):
    for p in ax.patches:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=10)

# Contagem do Impacto da Falta de Água
impact_count = df['Impacto da Falta de Água'].value_counts()

# Gráfico de Barra do Impacto da Falta de Água
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=impact_count.index, y=impact_count.values, palette='viridis')
add_value_labels(ax)
plt.title('Impacto da Falta de Água', fontsize=16)
plt.xlabel('Impacto', fontsize=14)
plt.ylabel('Contagem', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

# Contagem das Principais Dificuldades
dificuldade_count = df['Principal Dificuldade'].value_counts()

# Gráfico de Barra das Principais Dificuldades
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=dificuldade_count.index, y=dificuldade_count.values, palette='plasma')
add_value_labels(ax)
plt.title('Principais Dificuldades Durante a Falta de Água', fontsize=16)
plt.xlabel('Dificuldade', fontsize=14)
plt.ylabel('Contagem', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

# Contagem da Utilidade das Soluções
solucoes_count = df['Utilidade das Soluções'].value_counts()

# Gráfico de Barra da Utilidade das Soluções
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=solucoes_count.index, y=solucoes_count.values, palette='inferno')
add_value_labels(ax)
plt.title('Utilidade das Soluções Propostas', fontsize=16)
plt.xlabel('Utilidade', fontsize=14)
plt.ylabel('Contagem', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

# Contagem da Melhora da Situação
melhora_count = df['Melhora da Situação'].value_counts()

# Gráfico de Barra da Melhora da Situação
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=melhora_count.index, y=melhora_count.values, palette='magma')
add_value_labels(ax)
plt.title('Melhora da Situação com as Soluções Propostas', fontsize=16)
plt.xlabel('Melhora', fontsize=14)
plt.ylabel('Contagem', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

# Análise de Comentários
print("\nComentários dos Participantes:")
for comentario in df['Comentários']:
    print(f"- {comentario}")
```
